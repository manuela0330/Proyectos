Ahora se debe hallar la ascensión recta y declinación. Las coordenadas que obtenemos del momento minimo entre Apophis-tierra estan en cartesianas eclipticas y se pasarán ecuatoriales por medio de una matriz de rotacion respecto al eje x ya que la oblicuidad de la eclptica esta inclinada respecto al eje x.

\begin{equation}
  R_x(ɛ) =
\begin{bmatrix}
1 & 0 & 0 \\
0 & \cos\varepsilon & -\sin\varepsilon \\
0 & \sin\varepsilon & \cos\varepsilon
\end{bmatrix}
\end{equation}

```python
vector_apophis = rs[9, indice_minimo]

def matriz_rotacion(vector):
  e = np.radians(23.43) #oblicuidad de la ecliptica
  R = np.array([
      [1, 0, 0],
      [0, np.cos(e), -np.sin(e)],
      [0, np.sin(e), np.cos(e)]
  ])

  return np.dot(R,vector)


vector_eclipticas = matriz_rotacion(vector_apophis)
```

Teniendo ahora x,y,z en ecuatoriales se puede hallar la Ascensión recta y Declinación por medio de las ecuaciones

\begin{equation}
  \delta = \sin^{-1}\left(\frac{z}{r}\right)
\end{equation}

\begin{equation}
  \alpha = \tan^{-1}\left(\frac{y}{x}\right)
\end{equation}


```python
x,y,z = vector_eclipticas
r = np.linalg.norm(vector_eclipticas)

dec = np.arcsin(z/r)

ar = np.arctan2(y,x)
if ar < 0:
  ar = ar + 2*np.pi

ar_deg = np.degrees(ar)
dec_deg = np.degrees(dec)
ar_hora = Angle(ar_deg * u.deg)
```

en este caso, se utiliza una funcion de Astropy para pasar de grados a horas,minutos y segundos para que quede mejor reportada la ascesión recta (α)

```python
print(f'la ascensión recta en coordenadas ecuatoriales es:',
      ar_hora.to_string(unit=u.hour, sep=':', precision=2))

print(f'la declinación en coordenadas ecuatoriales es: {dec_deg:.2f}°')
```

```python
orbit = simulacion.particles[1].orbit()

#se integra 0.4yr
simulacion.integrate(0.4)

orbit_final = simulacion.particles[1].orbit()


def mostrar_elementos(orbit):

  a =  orbit.a
  e = orbit.e
  inc = orbit.inc
  Omega = orbit.Omega
  omega = orbit.omega
  f = orbit.f
  p = a * (1 - e**2)

  print(f" a = {orbit.a:.6f} au")
  print(f" e = {orbit.e:.6f} ")
  print(f" = {np.degrees(orbit.inc):.3f}°")
  print(f" Ω = {np.degrees(orbit.Omega):.3f}°")
  print(f" ω = {np.degrees(orbit.omega):.3f}°")
  print(f" f = {np.degrees(orbit.f):.3f}°\n")
  print(f" p = {p:.6f} au")


print("=== Elementos iniciales (antes del encuentro) ===")
mostrar_elementos(orbit)

print("=== Elementos finales (después del encuentro) ===")
mostrar_elementos(orbit_final)
```

en este caso, se tuvo que poner una función en la que se asegure que la anomalia media este entre 0 y 2Π ya que si no, la funcion no converge

```python
x_posicion = []
y_posicion = []

for t in fechas_apo:
  M_fin = M0_apo + n*(t - t_inicial)
  M_mod = np.mod(M_fin, 2*np.pi)
  F = newton(lambda F: e_apo*np.sinh(F) - F - M_mod, M_mod)
  f = 2 * np.arctan(np.sqrt((e_apo + 1)/(e_apo - 1))*np.tanh(F/2))

  r = p_apo / (e_apo*np.cosh(F) -1)
  x = r * np.cos(f)
  y = r * np.sin(f)

  #se rota a sistema ecliptico
  rot = rot_tp @ np.array([x,y,0])
  x_posicion.append(rot[0])
  y_posicion.append(rot[1])


plt.figure(figsize = (10,5))
plt.plot(x_posicion, y_posicion, label = 'Órbita apophis geocentrica')
plt.xlabel('x [AU]')
plt.ylabel('y [AU]')
plt.title('Órbita de Apophis geocentrica')
plt.grid()
plt.legend()
```

```python
def dibuja_conica(p=1.0, e=0.0):
  # Propiedades derivadas
  factor = 0.98
  if e > 1:
    psi = np.arccos(1/e)
    teta_min = -(np.pi-psi)*factor
    teta_max = (np.pi-psi)*factor

  fs = np.linspace(teta_min, teta_max, 100)
  rs = p/(1+e*np.cos(fs))

  xs = rs * np.cos(fs)
  ys = rs * np.sin(fs)

  plt.plot(xs, ys)
  plt.plot(0, 0, 'bo', label='Tierra')
  plt.title('Órbita de Apophis Geocentrica')
  plt.xlabel('x [AU]')
  plt.ylabel('y [AU]')
  plt.axis('equal')
  plt.grid()
  plt.show()


dibuja_conica(p1, e1)
```
